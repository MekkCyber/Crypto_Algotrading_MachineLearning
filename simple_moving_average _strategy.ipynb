```python
btc = yf.download('BTC-USD',start='2014-09-17',end='2023-10-22')
btc
```

```python
btc.Close.plot(figsize=(12,8))
```

```python
btc['returns'] = (btc['Close'].sub(btc['Close'].shift(1))).div(btc['Close'].shift(1)).mul(100)
btc['Change%'] = btc['Close'].pct_change(periods=1)
btc.dropna(inplace=True)
btc
```

```python
btc['returns'].plot(kind='hist',bins=100, figsize=(12,8))
```

```python
btc.returns.mean(), btc.returns.std()
# returns are often misleading that is why we use log returns
```

```python
def get_mean_std(df) : 
    new_df = df.dropna()
    new_df['% Change'] = new_df.pct_change(periods=1).mul(100)
    new_df.dropna(inplace=True)
    return new_df['% Change'].mean(), new_df['% Change'].std()

X = []
y = []
for ticker in cryptos.columns : 
    mean, std = get_mean_std(cryptos[ticker])
    X.append(std)
    y.append(mean)
plt.figure(figsize=(20,12))
plt.scatter(X,y, s=50)
for i in range(len(cryptos.columns)):
    plt.annotate(cryptos.columns[i].split('-')[0],xy=(X[i]+0.002,y[i]+0.002),size=15)
plt.xlabel("Annual risk(std)",fontsize = 15)
plt.ylabel("Annual return",fontsize = 15)
plt.title("Risk/return",fontsize=25)
plt.show()
```

### Log Returns

```python
btc['Log Returns'] = np.log(btc['Close']/btc['Close'].shift(1)).dropna()
btc['Cum Return'] = btc['Log Returns'].cumsum().apply(np.exp)
btc.dropna(inplace=True)
```

### Cumulative Return and drawdowns

```python
btc.loc[:,'log_returns'] = np.log(btc['Close'].div(btc['Close'].shift(1)))
btc.dropna(inplace=True)
```

```python
btc.loc[:,'cumulreturn'] = btc['log_returns'].cumsum().apply(np.exp)
btc['cumulreturn'].plot(figsize=(12,8),title="APPLe buy and hold", fontsize=12)
```

```python
btc.loc[:,"cummmax"]=btc.cumulreturn.cummax()
```

```python
btc[["cumulreturn","cummmax"]].plot(figsize=(12,8),title="APPLe buy and hold", fontsize=12)
```

```python
btc.loc[:,"drawdown"] = btc["cummmax"] - btc["cumulreturn"]
btc["drawdown"].plot(figsize=(12,8),title="APPLe buy and hold", fontsize=12)
```

```python
# very high drawdown
# the drawdown at a certain point of time, is the difference between the highest cumulative return till the fixed point of time and the current 
# cumulative return
btc.drawdown.max()
```

```python
def test_strategy(crypto, SMA, adjusted_for_long_bias=True):
    crypto_name = crypto.name
    crypto = crypto.dropna()
    crypto = pd.DataFrame(crypto)
    crypto.columns = ['Close']
    print(crypto.columns)
    crypto["returns"]=np.log(crypto.div(crypto.shift(1)))
    crypto["SMA_S"]=crypto.Close.rolling(int(SMA[0])).mean()
    crypto["SMA_L"]=crypto.Close.rolling(int(SMA[1])).mean()
    crypto.dropna(inplace=True)
    crypto["position"]=np.where(crypto["SMA_S"]>crypto["SMA_L"],1,0 if adjusted_for_long_bias else -1)
    crypto["strategy"]=crypto["returns"]*crypto.position.shift(1)
    crypto.dropna(inplace=True)
    
    std=crypto["strategy"].std()*np.sqrt(252)

    crypto["returnsbh"]=crypto["returns"].cumsum().apply(np.exp)
    crypto["returnsstrategy"]=crypto["strategy"].cumsum().apply(np.exp)
    perf=crypto["returnsstrategy"].iloc[-1]
    title="{}| SMA_S={} | SMA_L{}".format(crypto_name,SMA[0], SMA[1])
    crypto[["returnsbh","returnsstrategy"]].plot(title=title, figsize=(12,8))
    plt.xticks(rotation=30)
    return perf, std

perf, std = test_strategy(cryptos['BTC-USD'], [50,200])
print("performance : ", perf, " std : ",std)
```
